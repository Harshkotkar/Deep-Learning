```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import string

nltk.download('stopwords', quiet=True)
nltk.download('punkt', quiet=True)

sentence_A = "natural language processing enables machines to understand human language"
sentence_B = "machine understandd human language using nlp techniques"
sentence_C = "deep learning models are effective in image recognization"


def preprocess_sentence(sentence):
    # Convert to lowercase
    sentence = sentence.lower()
    # Remove punctuation
    sentence = sentence.translate(str.maketrans('', '', string.punctuation))
    # Tokenize words
    words = word_tokenize(sentence)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)

preprocessed_A = preprocess_sentence(sentence_A)
preprocessed_B = preprocess_sentence(sentence_B)
preprocessed_C = preprocess_sentence(sentence_C)

print(f"Preprocessed A: {preprocessed_A}")
print(f"Preprocessed B: {preprocessed_B}")
print(f"Preprocessed C: {preprocessed_C}")
```

```python
vectorizer = TfidfVectorizer()

sentences = [preprocessed_A, preprocessed_B, preprocessed_C]
tfidf_matrix = vectorizer.fit_transform(sentences)

# Calculate cosine similarity
similarity_AB = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
similarity_AC = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[2:3])[0][0]
similarity_BC = cosine_similarity(tfidf_matrix[1:2], tfidf_matrix[2:3])[0][0]

print(f"Cosine Similarity between A and B: {similarity_AB:.4f}")
print(f"Cosine Similarity between A and C: {similarity_AC:.4f}")
print(f"Cosine Similarity between B and C: {similarity_BC:.4f}")
```

```python
sentence_E = "natural language processing is a powerful tool to understand text"
sentence_F = "nlp is a strong method for understanding text"

# Preprocess new sentences
preprocessed_E = preprocess_sentence(sentence_E)
preprocessed_F = preprocess_sentence(sentence_F)

print(f"Preprocessed E: {preprocessed_E}")
print(f"Preprocessed F: {preprocessed_F}")
```

```python
all_sentences = [preprocessed_A, preprocessed_B, preprocessed_C, preprocessed_E, preprocessed_F]
new_tfidf_matrix = vectorizer.fit_transform(all_sentences)

# Calculate pairwise cosine similarities
similarity_AE = cosine_similarity(new_tfidf_matrix[0:1], new_tfidf_matrix[3:4])[0][0]
similarity_AF = cosine_similarity(new_tfidf_matrix[0:1], new_tfidf_matrix[4:5])[0][0]
similarity_BE = cosine_similarity(new_tfidf_matrix[1:2], new_tfidf_matrix[3:4])[0][0]
similarity_BF = cosine_similarity(new_tfidf_matrix[1:2], new_tfidf_matrix[4:5])[0][0]
similarity_CE = cosine_similarity(new_tfidf_matrix[2:3], new_tfidf_matrix[3:4])[0][0]
similarity_CF = cosine_similarity(new_tfidf_matrix[2:3], new_tfidf_matrix[4:5])[0][0]
similarity_EF = cosine_similarity(new_tfidf_matrix[3:4], new_tfidf_matrix[4:5])[0][0]

print(f"Cosine Similarity between A and E: {similarity_AE:.4f}")
print(f"Cosine Similarity between A and F: {similarity_AF:.4f}")
print(f"Cosine Similarity between B and E: {similarity_BE:.4f}")
print(f"Cosine Similarity between B and F: {similarity_BF:.4f}")
print(f"Cosine Similarity between C and E: {similarity_CE:.4f}")
print(f"Cosine Similarity between C and F: {similarity_CF:.4f}")
print(f"Cosine Similarity between E and F: {similarity_EF:.4f}")
```
