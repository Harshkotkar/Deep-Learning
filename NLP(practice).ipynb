```python
import requests #reads the messy HTML and makes it easy to search
from bs4 import BeautifulSoup
```

```python
url = "https://www.bbc.com/news/world"
response=requests.get(url)
```

```python
html=response.text
```

```python
soup=BeautifulSoup(response.text,'lxml')
```

```python
soup=BeautifulSoup(html,"html.parser")
```

```python
title=soup.find('title')
title
```

```python
t=soup.get_text(separator=' ',strip=True)
t
```

```python
import string
translator=str.maketrans(' ', ' ',string.punctuation)
s=t.translate(translator)
s=s[:300]
s
```

```python
import spacy
nlp=spacy.load("en_core_web_sm")
text= "the children. are running, in the #garden."
doc=nlp(text)
```

```python
filtered = [token.text for token in doc if not token.is_stop]
```

```python
s1=" ".join(filtered)
print(s1)
```

```python
new='red fox $jumps over @ the flex, and fall into #bucket.'
doc=nlp(new)
clean=[token.text for token in doc if not token.is_stop and token.is_alpha]
print(clean)
```

```python
new_clean=[token.text for token in doc if not token.is_stop]
print(new_clean)
```

```python
new_cl=[token.text for token in doc if token.is_alpha]
print(" ".join(new_cl))
```
