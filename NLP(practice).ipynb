```python
import requests #reads the messy HTML and makes it easy to search
from bs4 import BeautifulSoup
```

```python
url = "https://www.bbc.com/news/world"
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text,'lxml')
```

```python
print(response.text)
```

```python
title=soup.find('title')
title
```

```python
soup.find_all('a')
```

```python
quotes=soup.find_all('div',class_="quotes")
for i, quote in enumerate(quotes[:10],start=1):
  print("top 10 quotes",quote.get_text(strip=True))
```
