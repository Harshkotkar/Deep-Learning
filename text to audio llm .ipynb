```python
import sys
sys.path.append('/content/neutts-air/neuttsair')  # Adjust path

from neutts import NeuTTSAir
import soundfile as sf
import os

# Create samples directory if it doesn't exist
os.makedirs("samples", exist_ok=True)

# Create a sample reference text file
with open("samples/dave.txt", "w") as f:
    f.write("Hello, my name is Dave. I work as a software engineer.")

# Initialize TTS
tts = NeuTTSAir(
    backbone_repo="neuphonic/neutts-air-q4-gguf",
    backbone_device="cpu",
    codec_repo="neuphonic/neucodec",
    codec_device="cpu"
)

# For now, let's use text-to-speech without reference audio
# (since we don't have the actual dave.wav file)
input_text = "My name is Dave, and um, I'm from London."

# If you want to use a default voice, you might need to check the documentation
# for how to use the model without reference audio, or generate a simple voice
try:
    # Try inference without reference (if supported)
    wav = tts.infer(input_text)
    sf.write("test.wav", wav, 24000)
    print("Audio generated successfully as test.wav")
except Exception as e:
    print(f"Error: {e}")
    print("The model might require reference audio. You'll need to provide a .wav file.")
```
